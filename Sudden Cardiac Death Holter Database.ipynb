```python
import wfdb

# Specify the record name and path (excluding the file extension)
record_name = '/content/30'  # Adjust the path to where your .dat file is saved

# Read the record using wfdb
record = wfdb.rdrecord(record_name)

# Access the signal data
signal = record.p_signal

# Display information about the record
print(record.__dict__)

# Plot the ECG signal
wfdb.plot_wfdb(record=record, title='ECG Recording')
```
